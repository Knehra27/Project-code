```python
def create_siamese_pairs(x_train, y_train, num_pairs):
    pairs = []
    labels = []

    class_indices = defaultdict(list)
    for i, label in enumerate(y_train):
        class_indices[label].append(i)

    for _ in range(num_pairs // 2):
        pos_class = np.random.choice(list(class_indices.keys()))
        neg_class = np.random.choice(list(class_indices.keys()))
        while neg_class == pos_class:
            neg_class = np.random.choice(list(class_indices.keys()))

        i, j = np.random.choice(class_indices[pos_class], 2, replace=False)
        pairs.append([x_train[i], x_train[j]])
        labels.append(0)

        i = np.random.choice(class_indices[pos_class])
        j = np.random.choice(class_indices[neg_class])
        pairs.append([x_train[i], x_train[j]])
        labels.append(1)

    return np.array(pairs), np.array(labels).astype("float32")
```

```python
def euclidean_distance(vects):
    x, y = vects
    sum_square = K.sum(K.square(x - y), axis=1, keepdims=True)
    return K.sqrt(K.maximum(sum_square, K.epsilon()))
```

```python
def create_supp_set(imgs,lbls,shots):
  supp_im=[]
  supp_set_lbl=[]
  index=[]
  shift=[]
  shifted_lbl=[]
  i=0
  j=0
  tot=0
  for k in np.unique(lbls):
    n=0
    if k!=0:
      i+=1
    class_imgs = [img for img, lbl in zip(imgs, lbls) if lbl == k]
    shift.append(len(class_imgs))
    while n<shots:
      ind = random.randrange(len(class_imgs))
      while ind in index:
        ind = random.randrange(len(class_imgs))
      supp_im.append(class_imgs[ind])
      supp_set_lbl.append(k)
      index.append(ind)
      if k==0:
        shifted_lbl.append(ind)
      else:
        while j<k:
          tot+= shift[j]
          j+=1
        shifted_lbl.append(tot +ind)
      n+=1

  return supp_im,supp_set_lbl,index,shifted_lbl

```

```python
import numpy as np
import itertools

def create_siamese_pairs(x_train, y_train, num_pairs):
    """
    Create balanced Siamese pairs from the given images and labels.

    Args:
        x_train (np.ndarray): Array of images.
        y_train (np.ndarray): Array of labels.
        num_pairs (int): Number of pairs to generate.

    Returns:
        np.ndarray, np.ndarray: Arrays of image pairs and their labels.
    """
    pairs = []
    labels = []

    # Dictionary to hold indices for each class
    class_indices = {}
    for i in range(len(y_train)):
        label = y_train[i]
        if label not in class_indices:
            class_indices[label] = []
        class_indices[label].append(i)

    # Generate positive pairs
    positive_pairs = []
    for indices in class_indices.values():
        positive_pairs.extend(itertools.combinations(indices, 2))
    for (i, j) in positive_pairs:
        if len(pairs) < num_pairs // 2:
            pairs.append([x_train[i], x_train[j]])
            labels.append(0)

    num_pos_pairs = len(labels)
    # Generate negative pairs
    classes = list(class_indices.keys())
    while len(labels) < num_pairs:
        # Randomly select two different classes
        class1, class2 = np.random.choice(classes, size=2, replace=False)
        i = np.random.choice(class_indices[class1])
        j = np.random.choice(class_indices[class2])
        pairs.append([x_train[i], x_train[j]])
        labels.append(1)

    # Convert pairs and labels to numpy arrays
    pairs = np.array(pairs)
    labels = np.array(labels).astype("float32")

    return pairs, labels
```
